```python
# combine all filters together
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2019) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'BRA')]
```

```python
df2.head()
```

```python
#drop columns Athlete club, Athlete country, Athlete year of birth, Athlete age category
```

```python
df2=df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
#clean null values
```

```python
df2.isna().sum()
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#change column names
```

```python
df2 = df2.rename(columns = {'Year of event':'year',
                            'Event dates':'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_finishers',
                            'Athlete performance':'athlete_performance',
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id',
                            'Athlete age category': 'athlete_age_category',
                           })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3=df2[['race_day','race_name','race_length','race_number_finishers','athlete_id','athlete_gender','athlete_age_category', 'athlete_performance','athlete_average_speed','year']]
```

```python
df3.head()
```

```python
df3=df3.drop(['year'], axis=1)
df3.head()
```

```python
#graphics and charts
```

```python
sns.histplot(df3, x = 'race_length', hue = 'athlete_gender')
```

```python
sns.displot(df3[df3['race_length'] == '50km']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue = 'athlete_gender', split=True, inner='quart')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df3, x='athlete_age_category', y='athlete_average_speed', hue='athlete_age_category')
```

```python
df3.groupby(['athlete_age_category', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
df3.head()
```

```python
#seascons for data -> Slower in summer than winter?
# summer 11-2
# fall 3-5
#winter 6-8
#spring 9-11
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'summer' if x > 11 else 'spring' if x > 8 else 'winter' if x > 5 else 'fall' if x > 2 else 'summer')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed']. agg(['mean', 'count']).sort_values('mean', ascending=False)
```
